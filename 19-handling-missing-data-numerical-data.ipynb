```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('Datasets/titanic_toy.csv')
df
```

```python
df.isnull().mean()*100
```

```python
X= df.drop(columns=['Survived'])
y= df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape, X_test.shape
```

```python
X_train.isnull().mean()*100
```

```python
mean_age = X_train['Age'].mean()
median_age = X_train['Age'].median()

mean_fare = X_train['Fare'].mean()
median_fare = X_train['Fare'].median()
```

```python
X_train['Age_mean'] = X_train['Age'].fillna(mean_age)
X_train['Age_median'] = X_train['Age'].fillna(median_age)

X_train['Fare_mean'] = X_train['Fare'].fillna(mean_fare)
X_train['Fare_median'] = X_train['Fare'].fillna(median_fare)
```

```python
X_train.sample(50)
```

```python
print("Orignal Age variance:", X_train['Age'].var())
print("Age variance after mean imputation:", X_train['Age_mean'].var())
print("Age variance after median imputation:", X_train['Age_median'].var())
print("\n")
print("Orignal Fare variance:", X_train['Fare'].var())
print("Fare variance after mean imputation:", X_train['Fare_mean'].var())
print("Fare variance after median imputation:", X_train['Fare_median'].var())
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

X_train['Age'].plot(kind = 'kde', ax=ax, color='red')
X_train['Age_mean'].plot(kind = 'kde', ax=ax, color='orange')
X_train['Age_median'].plot(kind = 'kde', ax=ax, color='green')

lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

X_train['Fare'].plot(kind = 'kde', ax=ax, color='red')
X_train['Fare_mean'].plot(kind = 'kde', ax=ax, color='orange')
X_train['Fare_median'].plot(kind = 'kde', ax=ax, color='green')

lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
imputer1 = SimpleImputer(strategy='median')
imputer2 = SimpleImputer(strategy='mean')
```

```python
trf = ColumnTransformer([
    ('imputer1',imputer1,['Age']),
    ('imputer2',imputer2,['Fare'])
],remainder='passthrough')
```

```python
trf.fit(X_train)
```

```python
X_train = trf.transform(X_train)
X_test = trf.transform(X_test)
```
